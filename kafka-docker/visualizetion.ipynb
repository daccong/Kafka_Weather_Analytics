```python
import matplotlib.pyplot as plt
from datetime import datetime


# vẽ biểu đồ 
def plot_weather(data):
    datetim = data['days'][0]['datetime']
    hours = data['days'][0]['hours']
    times = [datetime.strptime(hour['datetime'], "%H:%M:%S").strftime("%H:%M") for hour in hours]
    temps = [hour['temp'] for hour in hours]
    humidities = [hour['humidity'] for hour in hours]
    plt.figure(figsize=(12, 6))
    plt.plot(times, temps, marker='o', label='Nhiệt độ (°F)', color='red')
    plt.plot(times, humidities, marker='s', label='Độ ẩm (%)', color='blue')
    plt.xticks(rotation=45)
    plt.xlabel("Thời gian")
    plt.ylabel("Giá trị")
    plt.title(f"Biểu đồ Nhiệt độ và Độ ẩm theo Giờ trong ngày {datetim}")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
plot_weather(df_hn)
```
